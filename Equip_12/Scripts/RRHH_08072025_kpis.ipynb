```python
import pandas as pd
```

```python
RRHH_08072025 = pd.read_csv(r"C:\Users\MAITE\Desktop\JOSE\IT_ACADEMY\2_ Especializacion Analisis de datos\x_Proyecto_Simulacion_Empresarial\Proyecto_RRHH\SPRINT_2\rrhh_08072025.csv")

RRHH_08072025 = RRHH_08072025.drop('Unnamed: 0', axis=1)
```

```python
RRHH_08072025.info()
```

```python
df_rrhh_08072025 = RRHH_08072025.copy()
```

```python
# Modificación de tipos de datos

columnas_a_convertir_1 = ['ID','Reason_absence','Month_absence','Day_week','Seasons','Disciplinary_failure','Education','Social_drinker','Social_smoker']
df_rrhh_08072025[columnas_a_convertir_1] = df_rrhh_08072025[columnas_a_convertir_1].apply(lambda col: col.astype('category'))

columnas_a_convertir_2 = ['Son','Pet']
df_rrhh_08072025[columnas_a_convertir_2] = df_rrhh_08072025[columnas_a_convertir_2].apply(lambda col: col.astype('int64'))


df_rrhh_08072025['Work_load_Average_day'] = pd.to_numeric(
    df_rrhh_08072025['Work_load_Average_day'].apply(lambda x: str(x).replace(',', '.') if isinstance(x, str) else x),
    errors='coerce').astype('float')

df_rrhh_08072025.info()
```

```python
# CREACION COLUMNAS NOMBRES DE CODIGOS TEMPORALES

# creacion columnas nombre estaciones
mapa = {1: 'summer', 2: 'autum', 3: 'winter',4:'spring'}
df_rrhh_08072025['Seasons_name'] = df_rrhh_08072025['Seasons'].map(mapa)

# # creacion columnas nombre meses
mapa2 = {0:'Unknown', 1:'January', 2:'February', 3:'March', 4:'April', 5:'May', 6:'June', 7:'July', 8:'August', 9:'September', 10:'October', 11:'November', 12:'December'}
df_rrhh_08072025['Month_absence_name'] = df_rrhh_08072025['Month_absence'].map(mapa2)

# creacion columnas nombre dias de semana
mapa3 = {2: 'Monday', 3: 'Tuesday',4:'Wednesday',5: 'Thursday', 6: 'Friday'}
df_rrhh_08072025['Day_week_name'] = df_rrhh_08072025['Day_week'].map(mapa3)

# cambio del tipo de datos
columnas_a_convertir = ['Seasons_name','Month_absence_name','Day_week_name']
df_rrhh_08072025[columnas_a_convertir] = df_rrhh_08072025[columnas_a_convertir].astype('category')
```

```python
df_rrhh_08072025
```

```python
# Ordenacion de la columna categorica de mes por nombre
ordered_month = ['January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December', 'Unknown']

df_rrhh_08072025['Month_absence_name'] = pd.Categorical(df_rrhh_08072025['Month_absence_name'], categories=ordered_month, ordered=True)

df_rrhh_08072025 = df_rrhh_08072025.sort_values('Month_absence_name')
df_rrhh_08072025
```

```python
# Calculo de la TASA MENSUAL DE ABSENTISMO

# Calculo del total de horas de absentismo por mes
tasa_mensual_absentismo = df_rrhh_08072025.groupby('Month_absence_name',observed=True)['Absenteeism_hours'].sum().reset_index()

tasa_mensual_absentismo.rename(columns={'Absenteeism_hours':'Monthly_Absenteeism_hours'}, inplace=True)

# Calculo de la tasa mensual de absentismo per capita (hora)
tasa_mensual_absentismo['Monthly_Absenteeism_rate'] = (tasa_mensual_absentismo['Monthly_Absenteeism_hours']/df_rrhh_08072025['ID'].nunique()).round(1)
tasa_mensual_absentismo = tasa_mensual_absentismo
tasa_mensual_absentismo
```

```python
# Calculo del MES PICO DE CARGA POR AUSENCIAS
indice_maximo = tasa_mensual_absentismo['Monthly_Absenteeism_hours'].idxmax()
mes_pico = tasa_mensual_absentismo.loc[indice_maximo, 'Month_absence_name']
mes_pico
```

```python
# # Calculo del MOTIVO MODAL DE ABSENTISMO
dic_reasons_absence = {22 :'patient follow-up', 23:'medical consultation', 28: 'dental consultation'}

reason_absence_count = df_rrhh_08072025['Reason_absence'].value_counts().reset_index()

index_max = reason_absence_count['count'].idxmax()
reason_absence_code = reason_absence_count.loc[index_max, 'Reason_absence']

motivo_modal_absentismo = dic_reasons_absence.get(reason_absence_code, 'Unknown Reason')

print(f'El motivo modal de absentismo laboral es "{motivo_modal_absentismo}"')
```

```python
# Calculo del INDICE DE EFICIENCIA RELATIVA POR CARGA LABORAL

indice_eficiencia_relativa_carga_laboral = (df_rrhh_08072025['Hit_target'].sum()/df_rrhh_08072025['Work_load_Average_day'].sum()).round(2)
print(f'El indice de eficiencia relativa por carga laboral es "{indice_eficiencia_relativa_carga_laboral}"')
```
