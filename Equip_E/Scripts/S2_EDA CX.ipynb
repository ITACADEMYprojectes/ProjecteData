### Gráfica de distribución normal para el rating

```python
# Calcular la media y desviación estándar de 'review_scores_rating'
media_rating = np.mean(df_with_rating['review_scores_rating'])
desviacion_rating = np.std(df_with_rating['review_scores_rating'])

# Crear un rango de valores para la curva normal
x = np.linspace(media_rating - 4*desviacion_rating, media_rating + 4*desviacion_rating, 1000)

# Generar la distribución normal (campana de Gauss)
y = stats.norm.pdf(x, media_rating, desviacion_rating)

plt.figure(figsize=(10, 5))

# Graficar la curva de la distribución normal
plt.plot(x, y, label=f'Curva Normal\nMedia: {media_rating:.2f}, Desviación Estándar: {desviacion_rating:.2f}', color='blue')

# Agregar línea vertical para la media
plt.axvline(x=media_rating, color='red', linestyle='--', label=f'Media: {media_rating:.2f}')

# Agregar líneas para la desviación estándar de 9.31 tanto en el lado positivo como negativo
linea_desviacion_pos = media_rating + 9.31  # 9.31 unidades por encima de la media
linea_desviacion_neg = media_rating - 9.31  # 9.31 unidades por debajo de la media

plt.axvline(x=linea_desviacion_pos, color='green', linestyle='--', label=f'+9.31 a la media: {linea_desviacion_pos:.2f}')
plt.axvline(x=linea_desviacion_neg, color='green', linestyle='--', label=f'-9.31 a la media: {linea_desviacion_neg:.2f}')

plt.legend(fontsize=8)
plt.title('Curva Normal de Review Scores Rating')
plt.xlabel('Review Scores Rating')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```

Con una media de 91,9 y una desviación estándar de 9, podemos concluir que los usuarios valoran en general el alojamiento de manera muy positiva. Sin embargo, la desviación estándar relativamente alta sugiere que existen algunas variaciones en la experiencia de los usuarios, lo que indica que mientras algunos usuarios tienen una experiencia excepcional, otros podrían haber tenido experiencias algo menos satisfactorias.


Las líneas de +9.31 y -9.31 marcan puntos extremos a esos valores de distancia de la media. Estas desviaciones ayudan a ver la dispersión de los datos y entender qué tan alejados estan de la media.

##  Prueba de Anderson-Darling

```python
result = stats.anderson(df_with_rating['review_scores_rating'], dist='norm')

print(f'Estadístico de prueba: {result.statistic:.3f}')
print(f'Valores críticos: {result.critical_values}')
print(f'Niveles de significancia: {result.significance_level}')

# Comparar el estadístico de prueba con los valores críticos
for i, sl in enumerate(result.significance_level):
    if result.statistic < result.critical_values[i]:
        print(f'Nivel de significancia {sl}%: No se rechaza la hipótesis nula (los datos son normales)')
    else:
        print(f'Nivel de significancia {sl}%: Se rechaza la hipótesis nula (los datos no son normales)')
```

La prueba indica que los datos no son normales.

### Prueba Mann-Whitney U
Dado que las variables no tienen una distribución normal, se realiza una prueba Mann-Whitney U, la más adecuada en este caso para determinar si la diferencia entre los dos grupos es estadísticamente significativa.

```python
stat, p_ac = mannwhitneyu(mejores['review_scores_accuracy'], peores['review_scores_accuracy'])
stat, p_ch = mannwhitneyu(mejores['review_scores_checkin'], peores['review_scores_checkin'])
stat, p_cl = mannwhitneyu(mejores['review_scores_cleanliness'], peores['review_scores_cleanliness'])
stat, p_co = mannwhitneyu(mejores['review_scores_communication'], peores['review_scores_communication'])
stat, p_lo = mannwhitneyu(mejores['review_scores_location'], peores['review_scores_location'])
stat, p_va = mannwhitneyu(mejores['review_scores_value'], peores['review_scores_value'])

print(f"Mann-Whitney U test p-value:")
print(f'Accuracy {p_ac}')
print(f'Check-in {p_ch}')
print(f'Cleanliness {p_ac}')
print(f'Communication {p_ac}')
print(f'Location {p_ac}')
print(f'Value {p_ac}')
```

El valor p es 0 (o prácticamente) para las 6 variables analizadas, por lo que la diferencia entre los dos grupos es estadísticamente significativa.

# Conclusiones y propuestas de negocios

Ambos análisis realizados confirman que los aspectos en que más se diferencian los alojamientos mejor y peor valorados son el valor percibido, la limpieza y la precisión de las descripciones. Aunque la ubicación influye, es donde menos variación hay entre los mejores y peores alojamientos.

__Value__

El valor es la primera variable a mejorar.

Es posible que una de las razones por las que este item tenga una gran diferencia entre mejores y peores evaluados es que sea una categoría poco definida, que da más espacio a la subjetividad. Proporcionar una definición clara de qué se evalúa al poner una puntuación de valor podría ayudar a nuestros anfitriones a lograr calificaciones que reflejen mejor la realidad. 

En el análisis de correlación del precio y el valor, puede inferirse que a más valor, más expectativa de calidad tendrán los huéspedes. La puntuación de este item se podrá mejorar elevando la calidad de los alojamientos en todos los aspectos o ajustando los precios para que el cliente tenga la sensación de que el precio se ajustó a los servicios ofrecidos.

Otra forma de mejorar la percepción de valor podría ser ofrecer servicios adicionales o descuentos.

__Cleanliness__

La limpieza es otro aspecto que presenta gran diferencia. Aunque las putuaciones de nuestros alojamientos son excepcionales, este aspecto tiene posibilidad de mejora.

Sugerimos invertir en personal de limpieza adicional, a la vez que se informa a los anfitriones sobre los estándares de limpieza para que puedan cumplirlos mejor.

También pueden realizarse auditorías periódicas para asegurar que todos los alojamientos estén a la altura de las expectativas, llamando la atención a los alojamientos que tengan peores calificaciones en esta área.

__Accuracy__

Proponemos asesorar a los anfitriones en la mejora de las descripciones y fotografías, las expectativas deben alinearse con la realidad. Pueden ofrecerse directrices claras sobre cómo deben ser las descripciones y sobre la calidad de las fotografías para que reflejen fielmente la realidad del alojamiento a la vez que atraen clientes.

Se sugiere también implementar un proceso de verificación regular para asegurarse de que las descripciones sigan siendo precisas.
