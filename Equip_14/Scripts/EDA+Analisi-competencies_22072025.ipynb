```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

top_3_tots['ANY'] = top_3_tots['ANY'].astype(int)
top_3_50_tots = top_3_50_tots.rename(columns={'AÑO': 'ANY'})
top_3_50_tots['ANY'] = top_3_50_tots['ANY'].astype(int)

replacements = {
    "De atención al público/ trato a clientes": "Atenció al públic",
    "De trabajo en equipo": "Treball en equip",
    "De resolución de problemas (localización de problemas o fallos, análisis de sus causas y búsqueda de soluciones)": "Resolució de problemes",
    "De dirección": "Direcció",
    "Técnicas específicas del puesto de trabajo": "Tècniques específiques"
}
top_3_tots['COMPETENCIA'] = top_3_tots['COMPETENCIA'].replace(replacements)
top_3_50_tots['COMPETENCIA'] = top_3_50_tots['COMPETENCIA'].replace(replacements)

top_3_tots['GRUP'] = 'Top 3 (Serveis)'
top_3_50_tots['GRUP'] = 'Top 3 (Tamany)'

df_comb = pd.concat([top_3_tots, top_3_50_tots], ignore_index=True)

plt.figure(figsize=(12, 6))
sns.lineplot(data=df_comb, x='ANY', y='MUCHO', hue='COMPETENCIA', style='GRUP', markers=True, dashes=False)

plt.title('Evolució de les competències millor valorades')
plt.xlabel('Any')
plt.ylabel('Percentatge')
plt.xticks(sorted(df_comb['ANY'].unique()))
plt.legend(title='Competència i grup', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
