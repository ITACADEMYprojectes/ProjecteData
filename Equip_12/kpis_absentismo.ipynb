```python
# MODIFICACION TIPO DE DATOS / DATOS DESCONOCIDOS      df_rrh_clean_1

df_rrhh_clean_1 = df_rrhh_clean_original.copy()

df_rrhh_clean_1['Reason_absence'].replace('motivo desconocido','Unknown',inplace=True)

df_rrhh_clean_1['Month_absence'] = df_rrhh_clean_1['Month_absence'].replace('mes desconocido','Unknown')


columnas_a_convertir = ['ID','Reason_absence','Month_absence','Day_week','Seasons','Disciplinary_failure','Education','Social_drinker','Social_smoker']
df_rrhh_clean_1[columnas_a_convertir] = df_rrhh_clean_1[columnas_a_convertir].astype('category')
df_rrhh_clean_1.info()
```

```python
# CREACION COLUMNAS NOMBRES DE CODIGOS TEMPORALES   df_rrhh_clean_2

df_rrhh_clean_2 = df_rrhh_clean_1.copy()

# creacion columnas nombre estaciones
mapa = {1: 'summer', 2: 'autum', 3: 'winter',4:'spring'}
df_rrhh_clean_2['Seasons_name'] = df_rrhh_clean_2['Seasons'].map(mapa)

# # creacion columnas nombre meses
mapa2 = {'Unknown' : 'Unknown','1': 'January', '2': 'February', '3': 'March','4':'April','5': 'May', '6': 'June', '7': 'July','8':'August', '9': 'September','10':'October', '11': 'November','12':'December'}
df_rrhh_clean_2['Month_absence_name'] = df_rrhh_clean_2['Month_absence'].map(mapa2)

# creacion columnas nombre dias de semana
mapa3 = {2: 'Monday', 3: 'Tuesday',4:'Wednesday',5: 'Thursday', 6: 'Friday'}
df_rrhh_clean_2['Day_week_name'] = df_rrhh_clean_2['Day_week'].map(mapa3)

# cambio del tipo de datos
columnas_a_convertir = ['Seasons_name','Month_absence_name','Day_week_name']
df_rrhh_clean_2[columnas_a_convertir] = df_rrhh_clean_2[columnas_a_convertir].astype('category')
```

```python
# CREACION DATAFRAMES EMPLOYEEES, WORK, ABSEN

df_rrhh_clean_3 = df_rrhh_clean_2.copy()

df_employees = df_rrhh_clean_3[['ID','Age','Weight','Height','Body_mass_index','Education','Son','Pet','Social_drinker','Social_smoker']].drop_duplicates().reset_index(drop=True)

df_work = df_rrhh_clean_3[['ID','Service_time','Distance_Residence_Work','Transportation_expense']].drop_duplicates().reset_index(drop=True)

df_absen = df_rrhh_clean_3[['ID','Reason_absence','Absenteeism_hours','Estimated_Absence_Days','Month_absence_name','Day_week_name','Seasons_name','Hit_target','Work_load_Average_day','Disciplinary_failure']]

```

```python
# CREACION DATAFRAME PKI

pki = df_absen.copy()

# Filtrado de los valores de mes
pki = pki[~ pki['Month_absence_name'].isin(['Unknown'])]

# Ordenacion de la columna categorica de mes por nombre
ordered_month = ['January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December']

pki['Month_absence_name'] = pd.Categorical(pki['Month_absence_name'], categories=ordered_month, ordered=True)

pki = pki.sort_values('Month_absence_name')

pki
```

```python
# Calculo de la tasa mensual de absentismo 

# Calculo del total de horas de absentismo por mes
tasa_mensual_absentismo = pki.groupby('Month_absence_name',observed=True)['Absenteeism_hours'].sum().reset_index()

tasa_mensual_absentismo.rename(columns={'Absenteeism_hours':'Monthly_Absenteeism_hours'}, inplace=True)

# Calculo de la tasa mensual de absentismo per capita (hora)
tasa_mensual_absentismo['Monthly_Absenteeism_rate'] = (tasa_mensual_absentismo['Monthly_Absenteeism_hours']/df_employees['ID'].nunique()).round(1)
tasa_mensual_absentismo
```

```python
# Calculo del mes pico de carga por ausencias
indice_maximo = tasa_mensual_absentismo['Monthly_Absenteeism_hours'].idxmax()
mes_pico = tasa_mensual_absentismo.loc[indice_maximo, 'Month_absence_name']
mes_pico
```

```python
# # Calculo del motivo modal de absentismo

dic_reasons_absence = {'22':'patient follow-up','23':'medical consultation','28': 'dental consultation'}

reason_absence_count = pki['Reason_absence'].value_counts().reset_index()

index_max = reason_absence_count['count'].idxmax()
reason_absence_code = reason_absence_count.loc[index_max, 'Reason_absence']

motivo_modal_absentismo = dic_reasons_absence.get(reason_absence_code, 'Unknown Reason')

print(f'El motivo modal de absentismo laboral es "{motivo_modal_absentismo}"')
```

```python
# Calculo del indice de eficiencia relativa por carga laboral

indice_eficiencia_relativa_carga_laboral = (pki['Hit_target'].sum()/pki['Work_load_Average_day'].sum()).round(2)
print(f'El indice de eficiencia relativa por carga laboral es "{indice_eficiencia_relativa_carga_laboral}"')
```

```python
# GRAFICOS

# Calcular la media de la columna 'Monthly_Absenteeism_rate'
mean_Monthly_Absenteeism_rate = tasa_mensual_absentismo['Monthly_Absenteeism_rate'].mean()

# Crear un lineplot
sns.lineplot(
    data=tasa_mensual_absentismo,              
    x="Month_absence_name",             
    y="Monthly_Absenteeism_rate",       
    marker = 'o',
    legend="auto",                  
    )

# Agregar una lÃ­nea horizontal con la media
plt.axhline(y=mean_Monthly_Absenteeism_rate, color='red', linestyle='--', label=f'Mean = {mean_Monthly_Absenteeism_rate:.2f}')

plt.title("Monthly_Absenteeism_rate")
plt.xlabel("MONTH")
plt.xticks(rotation=45)
plt.ylabel("RATE")
plt.grid(True)
plt.tight_layout()
plt.show()
```
