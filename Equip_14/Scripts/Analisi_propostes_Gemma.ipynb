```python
import matplotlib.pyplot as plt
import pandas as pd

def classificar_imc(imc):
    if imc < 18.5:
        return 'Pes insuficient'
    elif imc < 25:
        return 'Normal'
    elif imc < 30:
        return 'Sobrepès'
    elif imc < 35:
        return 'Obesitat lleu'
    else:
        return 'Obesitat moderada'

df_rrhh_nodup['Body_mass_index'] = df_rrhh_nodup['Body_mass_index'].apply(classificar_imc)

counts = df_rrhh_nodup['Body_mass_index'].value_counts().sort_index()
labels = counts.index
colors = ['lightblue', 'lightgreen', 'gold', 'orange', 'lightcoral']

plt.figure(figsize=(7, 7))
plt.pie(counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title("Distribució dels treballadors segons franges d'IMC")
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt

df_rrhh_nodup['Color'] = df_rrhh_nodup['Social_drinker'].map({True: 'purple', False: 'green'})
df_rrhh_nodup['Marker'] = df_rrhh_nodup['Social_smoker'].map({True: 'x', False: 'o'})

plt.figure(figsize=(10, 6))

for color in ['purple', 'green']:
    for marker in ['o', 'x']:
        subset = df_rrhh_nodup[(df_rrhh_nodup['Color'] == color) & (df_rrhh_nodup['Marker'] == marker)]
        plt.scatter(subset['Weight'], subset['Body_mass_index'],
                    c=color, marker=marker, label=f"{'Bevedor' if color == 'purple' else 'No bevedor'} {'i fumador' if marker == 'x' else 'i no fumador'}",
                    alpha=0.8, edgecolors='k')

plt.xlabel("Pes (kg)")
plt.ylabel("Índex de massa corporal (IMC)")
plt.title("IMC i pes segons hàbits socials (beure i fumar)")
plt.legend()
plt.show()

```
