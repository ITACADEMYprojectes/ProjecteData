```python
# Modificaci√≥n de tipos de datos


columnas_a_convertir_1 = ['ID','Reason_absence','Month_absence','Day_week','Seasons','Disciplinary_failure','Education','Social_drinker','Social_smoker']
df_rrhh_08072025[columnas_a_convertir_1] = df_rrhh_08072025[columnas_a_convertir_1].apply(lambda col: col.astype('category'))

columnas_a_convertir_2 = ['Son','Pet']
df_rrhh_08072025[columnas_a_convertir_2] = df_rrhh_08072025[columnas_a_convertir_2].apply(lambda col: col.astype('int64'))


df_rrhh_08072025['Work_load_Average_day'] = pd.to_numeric(
    df_rrhh_08072025['Work_load_Average_day'].apply(lambda x: str(x).replace(',', '.') if isinstance(x, str) else x),
    errors='coerce').astype('float')

df_rrhh_08072025
```
