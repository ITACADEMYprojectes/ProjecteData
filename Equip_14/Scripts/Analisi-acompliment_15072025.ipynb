```python
import pandas as pd

df_rrhh = pd.read_csv('/Users/gemmaseguesmajos/Downloads/DF-RRHH_15072025.csv', sep=',', encoding='utf-8')

df_rrhh.info()
```

```python
df_rrhh.isnull().sum()
```

```python
df_rrhh = df_rrhh.drop_duplicates()
df_rrhh = df_rrhh.sort_values(by='ID').reset_index(drop=True)
df_rrhh
```

```python
df_rrhh['ID'] = df_rrhh['ID'].astype(str).str.strip()
df_rrhh['Reason_absence'] = df_rrhh['Reason_absence'].astype(str).str.strip()
df_rrhh['Work_load_Average_day'] = df_rrhh['Work_load_Average_day'].astype(str).str.replace(',', '.', regex=False).astype(float)
df_rrhh['Education'] = df_rrhh['Education'].astype(str).str.strip()
df_rrhh['Social_drinker'] = df_rrhh['Social_drinker'].astype(int)
df_rrhh['Social_smoker'] = df_rrhh['Social_smoker'].astype(int)
df_rrhh['Disciplinary_failure'] = df_rrhh['Disciplinary_failure'].astype(int)
df_rrhh['Son'] = df_rrhh['Son'].astype(int)
df_rrhh['Pet'] = df_rrhh['Pet'].astype(int)

df_rrhh.info()
```

```python
df_rrhh["Disciplinary_failure"] = df_rrhh["Disciplinary_failure"].astype(bool)
df_rrhh["Social_drinker"] = df_rrhh["Social_drinker"].astype(bool)
df_rrhh["Social_smoker"] = df_rrhh["Social_smoker"].astype(bool)

df_rrhh.info()
```

```python
df_rrhh['Month_absence'].unique()
df_rrhh['Day_week'].unique()
```

```python
season_dict = {
    1: 'Hivern',
    2: 'Estiu',
    3: 'Tardor',
    4: 'Primavera'
}

df_rrhh['Seasons'] = df_rrhh['Seasons'].map(season_dict)
```

```python
df_rrhh
```

```python
month_dict = {
    1: 'Gen',
    2: 'Feb',
    3: 'Mar',
    4: 'Abr',
    5: 'Mai',
    6: 'Jun',
    7: 'Jul',
    8: 'Ago',
    9: 'Set',
    10: 'Oct',
    11: 'Nov',
    12: 'Des'
}

df_rrhh['Month_absence'] = df_rrhh['Month_absence'].map(month_dict)

df_rrhh
```

```python
day_dict = {
    2: 'Dll',
    3: 'Dm',
    4: 'Dc',
    5: 'Dj',
    6: 'Dv'
    }

df_rrhh['Day_week'] = df_rrhh['Day_week'].map(day_dict)

df_rrhh
```

```python
df_rrhh_nodup = df_rrhh.drop_duplicates(subset="ID")
df_rrhh_nodup = df_rrhh_nodup.sort_values(by='ID').reset_index(drop=True)
df_rrhh_nodup.index = df_rrhh_nodup.index + 1 

df_rrhh_nodup['ID'] = range(1, len(df_rrhh_nodup) + 1)

columns_to_keep = [
    "ID", "Transportation_expense", "Distance_Residence_Work",
    "Service_time", "Age", "Disciplinary_failure", "Education", "Son",
    "Social_drinker", "Social_smoker", "Pet", "Weight", "Height", "Body_mass_index"
]

df_rrhh_nodup["ID"] = df_rrhh_nodup["ID"].astype(str).str.strip()
df_rrhh_nodup = df_rrhh_nodup[columns_to_keep]
df_rrhh_nodup
df_rrhh_nodup
```

```python
from scipy.stats import f_oneway

group0 = df_rrhh[df_rrhh['Disciplinary_failure'] == 0]
group1 = df_rrhh[df_rrhh['Disciplinary_failure'] == 1]

anova_service = f_oneway(group0["Service_time"], group1["Service_time"])
anova_workload = f_oneway(group0["Work_load_Average_day"], group1["Work_load_Average_day"])
anova_distance = f_oneway(group0["Distance_Residence_Work"], group1["Distance_Residence_Work"])

print(f"ANOVA Service_time: F={anova_service.statistic:.3f}, p-value={anova_service.pvalue:.3f}")
print(f"ANOVA Work_load_Average_day: F={anova_workload.statistic:.3f}, p-value={anova_workload.pvalue:.3f}")
print(f"ANOVA Distance_Residence_Work: F={anova_distance.statistic:.3f}, p-value={anova_distance.pvalue:.3f}")
```

```python
import statsmodels.api as sm

df_filtrat = df_rrhh[df_rrhh['Absenteeism_hours'] < 8]

X = df_filtrat[['Age', 'Son', 'Pet', 'Service_time', 'Work_load_Average_day', 'Distance_Residence_Work']]
X = sm.add_constant(X)

y = df_filtrat['Absenteeism_hours']

model = sm.OLS(y, X).fit()

print(model.summary())
```

```python
import statsmodels.api as sm

X = df_rrhh[['Service_time', 'Work_load_Average_day', 'Distance_Residence_Work', 'Age', 'Son', 'Pet']]

X = sm.add_constant(X)  
y = df_rrhh['Disciplinary_failure']

model = sm.Logit(y, X)
result = model.fit()

print(result.summary())
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X, y)

print('Distribució original:')
print(y.value_counts())
print('Distribució després de SMOTE:')
print(pd.Series(y_res).value_counts())
```

```python
import statsmodels.api as sm

X_res = sm.add_constant(X_res)  
model = sm.Logit(y_res, X_res)
result = model.fit()
print(result.summary())
```
